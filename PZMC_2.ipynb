```python
import numpy as np
import pandas as pd
import statistics
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
import scipy.stats as stats

# Функции-----------------------------------------------------------------------

# Задание №1

def Values(data):
  mean_num = statistics.mean(data['number'])
  std_dev = statistics.pstdev(data['number'])

  return mean_num, std_dev

# Задание №2

def Laplas(data, mean_num, std_dev):
  n = len(data)

  # Уровень доверия
  confidence_level = 0.95
  # Критическое значение z для заданного уровня доверия
  z = stats.norm.ppf((1 + confidence_level) / 2)

  # Вычисление доверительного интервала
  margin_of_error = z * (std_dev / np.sqrt(n))
  confidence_interval = (round(mean_num - margin_of_error, 3), round(mean_num + margin_of_error, 3))

  return confidence_interval

# Задание №3

def Student(data, mean_num, std_dev):
  n = len(data)
  confidence_level = 0.95
  degrees_of_freedom = n - 1

  # Критическое значение t для заданного уровня доверия и степеней свободы
  t_critical = stats.t.ppf((1 + confidence_level) / 2, degrees_of_freedom)

  # Вычисление доверительного интервала
  margin_of_error = t_critical * (std_dev / np.sqrt(n))
  confidence_interval = (round(mean_num - margin_of_error, 3), round(mean_num + margin_of_error, 3))

  return confidence_interval

# Задание №4

def Square(data, mean_num, std_dev):
   # Вычисление выборочной дисперсии
    sample_variance = statistics.variance(data['number'])
    n = len(data)

    # Уровень доверия (например, 95%)
    confidence_level = 0.95
    alpha = 1 - confidence_level

    # Степени свободы
    degrees_of_freedom = n - 1

    # Критические значения χ² для заданного уровня доверия и степеней свободы
    chi2_lower = stats.chi2.ppf(alpha / 2, degrees_of_freedom)
    chi2_upper = stats.chi2.ppf(1 - alpha / 2, degrees_of_freedom)

    # Вычисление доверительного интервала для стандартного отклонения
    confidence_interval_lower = np.sqrt((degrees_of_freedom * sample_variance) / chi2_upper)
    confidence_interval_upper = np.sqrt((degrees_of_freedom * sample_variance) / chi2_lower)

    return (round(confidence_interval_lower, 3), round(confidence_interval_upper, 3))

# Загрузка данных---------------------------------------------------------------

data = pd.read_csv('1.txt', sep='\t', header=None, names=['number'])
data_2 = pd.read_csv('2.txt', sep='\t', header=None, names=['number'])
data_3 = pd.read_csv('3.txt', sep='\t', header=None, names=['number'])
data_4 = pd.read_csv('4.txt', sep='\t', header=None, names=['number'])

# Выполнение задания

# 1

mean_num_1, std_dev_1 = Values(data)
mean_num_2, std_dev_2 = Values(data_2)
mean_num_3, std_dev_3 = Values(data_3)
mean_num_4, std_dev_4 = Values(data_4)

# 2

confidence_interval_21 = Laplas(data, mean_num_1, std_dev_1)
confidence_interval_22 = Laplas(data_2, mean_num_2, std_dev_2)
confidence_interval_23 = Laplas(data_3, mean_num_3, std_dev_3)
confidence_interval_24 = Laplas(data_4, mean_num_4, std_dev_4)

# 3

confidence_interval_31 = Student(data, mean_num_1, std_dev_1)
confidence_interval_32 = Student(data_2, mean_num_2, std_dev_2)
confidence_interval_33 = Student(data_3, mean_num_3, std_dev_3)
confidence_interval_34 = Student(data_4, mean_num_4, std_dev_4)

# 4

confidence_interval_41 = Square(data, mean_num_1, std_dev_1)
confidence_interval_42 = Square(data_2, mean_num_2, std_dev_2)
confidence_interval_43 = Square(data_3, mean_num_3, std_dev_3)
confidence_interval_44 = Square(data_4, mean_num_4, std_dev_4)

#Итоги--------------------------------------------------------------------------

df = pd.DataFrame({
    '1.txt': [len(data), mean_num_1, std_dev_1, confidence_interval_21, confidence_interval_31, confidence_interval_41],
    '2.txt': [len(data_2), mean_num_2, std_dev_2, confidence_interval_22, confidence_interval_32, confidence_interval_42],
    '3.txt': [len(data_3), mean_num_3, std_dev_3, confidence_interval_23, confidence_interval_33, confidence_interval_43],
    '4.txt': [len(data_4), mean_num_4, std_dev_4, confidence_interval_24, confidence_interval_34, confidence_interval_44],
}, index=['Количество элементов', 'Среднее значение', 'Среднеквадратическое отклонение', 'Доверительного интервала мат. ожидания', 't-распределения Стьюдента', 'Доверительный интервал для среднеквадратического отклонения'])

display(df)
```
